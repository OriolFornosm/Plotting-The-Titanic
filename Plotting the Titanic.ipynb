# Plotting the Titanic

This notebook presents a practical exercise using Seaborn Lib for Python doing different types of plots based on the Titanic Data Set implemented in the Seaborn Lib itself.

### Importing the Lib

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Setting up the style of the plots

```python
sns.set_style('darkgrid')
```

### Loading the Data Set
And showing the head of the Data Frame

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Plots

A distribution plot that crosses two variables; age and fare paid.

```python
sns.set_context("notebook")
sns.jointplot(x="fare", y="age", data=titanic)
```

A plot showing the age distribution of the passengers.

```python
plt.figure(figsize=(12,10))
sns.distplot(titanic["age"], kde=False, bins=30)
```

First: A box type plot, showing crossing the age variable with the class wich passengers paid for.

Second: Same variables crossed but this time using a swarm plot, wich can be more presice as every dot is a data measure.

```python
plt.figure(figsize=(12,10))
sns.boxplot(x = "class", y = "age", data= titanic, palette ="rainbow")
```

```python
plt.figure(figsize=(12,10))
sns.swarmplot (x="class", y="age", data=titanic, palette = "RdPu_r")
```

A simple count plot, couting the number of passangers based on their gender.

```python
plt.figure(figsize=(10,10))
sns.countplot(x="sex", data=titanic, palette = "coolwarm")
```

A total correlation heat map, to show the consistency between all the different variables in the data set.

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(titanic.corr(), cmap="Blues")
plt.show()
```

Last plot, showing age distribution in two plots, one for males and the other for females.

```python
grid = sns.FacetGrid(titanic, col="sex")
grid = grid.map(plt.hist, "age")
```
